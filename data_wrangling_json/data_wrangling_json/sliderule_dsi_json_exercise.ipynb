## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# specify file path to world bank data
filepath = 'data/world_bank_projects.json'
```

```python
# create DataFrame from json file
# Each row of the DataFrame corresponds to a Country's project
# inspect accessible data
json_df = pd.read_json(filepath)
print(json_df.shape)
print(json_df.columns)
```

```python
# load json world bank data
wb_data = json.load(open(filepath))
```

```python
#create DataFrame from nested elements with desired columns
t = json_normalize(wb_data, 'mjtheme_namecode',['project_name','countryname' ])
t.columns = ['code', 'theme', 'project_name', 'countryname']
t = t.sort_values(by=['countryname'])
type(t)
t
```

```python
### PROBLEM 1. Find the 10 countries with most projects
```

```python
# return top 10 value_counts on t.countryname
df1 = t.countryname.value_counts()
print('Number of countries with projects: ', df1.shape)
df1.head(10)
```

```python
### PROBLEM 2. Find the top 10 major project themes (using column 'mjtheme_namecode')
```

```python
# print top 10 project themes
# head arg=11 because empty project themes were grouped and counted together
df2 = t.theme.value_counts()
df2.head(11)
```

```python
### PROBLEM 3. In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.
```

```python
df3 = t.sort_values(by=['code'])
df3
```

```python
# DataFrame of all rows with missing themes
empties = df3[df3.theme=='']
empties
```

```python
#Themes correspond to the code columns
#identify code: theme pairs
pairs = set(zip(df3.code, df3.theme))
pairs = [pair for pair in pairs if pair[1]!='']
pairs = dict(pairs)
pairs
```

```python
# map pairs to theme column ot fill missing values
df3.theme = df3.code.map(pairs)
```

```python
# An empty dataframe if all empty theme rows have been filled accordingly
empties2 = df3[df3.theme == '']
print(empties2.shape)
empties2
```

```python
# recaluculate solution to problem 2 with mapped themes
df3.theme.value_counts().head(10)
```
